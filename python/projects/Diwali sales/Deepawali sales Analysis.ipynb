Exploratory Data Analytics

Gender

```python
ax = sns.countplot(x = 'Gender',data=df)
ax
```

```python
ax = sns.countplot(x = 'Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

Females are done more expenditure as compared to men

AGE

```python
sns.countplot(x="Age Group", data=df,hue='Gender')
```

```python
ax = sns.countplot(x="Age Group", data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

AS per the above chart the most buyers are the 26-35 years womens

State

```python
#total numbers of orders from the top 10 state
sales_state_ord = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)

plt.figure(figsize=(15,15))
ax = sns.barplot(x="State",y="Orders",data=sales_state_ord)

for bars in ax.containers:
    ax.bar_label(bars)
```

Most of the orders from the uttar pradesh ,maharastra and karnataka

```python
df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
sales_state_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)

plt.figure(figsize=(15,15))
sns.barplot(x="State", y="Amount", data=sales_state_amt)
```

most of the amount generated by the states are uttar pradesh maharastra and karnataka

marital Status

```python
ax = sns.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#group by the  marital status and gender

sales_marry_gen = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x="Marital_Status", y="Amount", data=sales_marry_gen,hue="Gender")
```

Occupation

```python
plt.figure(figsize=(15,15))
ax = sns.countplot(x="Occupation", data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#group by the Occupation

df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending = False)
```

```python
sales_occup_amt = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,15))
ax = sns.barplot(x="Occupation", y="Amount",data=sales_occup_amt)

for bars in ax.containers:
    ax.bar_label(bars)
```

as per the graph the mot of the buyers are working in IT sector then Healthcare

Product Category

```python
#product category with the Amount
plt.figure(figsize=(25,10))
sns.countplot(x="Product_Category", data=df)
```

```python
#group by product category
df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_prod_cat_amt = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=(25,10))
ax = sns.barplot(x="Product_Category", y='Amount', data=sales_prod_cat_amt)

for bars in ax.containers:
    ax.bar_label(bars)
```

As per the chart the food and clothings are giving the more amount

```python
#By product id and orders

df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sales_prodid_orders = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,10))
sns.barplot(x='Product_ID',y='Orders',data=sales_prodid_orders)
```
